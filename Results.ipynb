```python
import matplotlib.pyplot as plt
import numpy
import os
import json
```

```python
MODELS_DIR_PATH = './models'
```

```python
loss_jsons = []
for file_name in os.listdir(MODELS_DIR_PATH):
    file_path = os.path.join(MODELS_DIR_PATH, file_name)
    try:
        if file_name.split('.')[1] == 'json':
            label_name = file_name.split('.')[0]
            loss_jsons.append((label_name, file_path))
    except:
        pass
for loss in loss_jsons:
    name = loss[0]
    if 'train' in name:
        line = json.load(open(loss[1]))
        plt.plot(line, label=name)
plt.ylim((0, 1))
plt.title('Train')
plt.legend()
plt.show()

for loss in loss_jsons:
    name = loss[0]
    if 'val' in name:
        line = json.load(open(loss[1]))
        plt.plot(line, label=name)
plt.title('Val')
plt.ylim((0, 5))
plt.legend()
plt.show()
```

Trained on 1000 epoch with `aaa6a05cc.tiff` as training image and `2f6ecfcdf.tiff` with `num_samples = 1000` and `batch_size = 100`

Based on val results:
1. **CBAM_R2UNet_v2**
  - Encoder: Recurrent Residual Block + CBAM
  - Decoder: R2UNet
  - UNet Skip Connections: Recurrent Residual Block + CBAM
2. **UNet**
  - Encoder: Conv2d Block
  - Decoder: Conv2d Block
  - UNet Skip Connections: Conv2d Block
3. **CBAM_R2UNet_v5**
  - Encoder: Recurrent Residual Block + CBAM
  - Decoder: Conv2d Block
  - UNet Skip Connections: Recurrent Residual Block + CBAM
